```python
import pandas as pd
import glob
import numpy as np
import random


all_data = pd.DataFrame()

input_files = glob.glob("../avg/ECU_DATA/*.csv")[-6:-1]
for file in input_files:
    print(file)
    df = pd.read_csv(file)

    # Battery voltage
    for index in range(1,len(df)):
        df["BatteryVoltage[BatteryVoltage]"][index] = random.randint(9000,12000)
    # MAPSource[Pressure]
        df["MAPSource[Pressure]"][index] = random.randint(470,600)
    # AirTemp[Temperature]
        df["AirTemp[Temperature]"][index] = random.randint(3800,4400)
    # CoolantTemp[Temperature]
        df["CoolantTemp[Temperature]"][index] = random.randint(3800,4700)
    # IgnitionAirTempCorrection[Angle]
        df["IgnitionAirTempCorrection[Angle]"][index] = random.randint(30,60)
    # IgnitionCoolantTempCorrection[AngleOffset10deg]
        df["IgnitionCoolantTempCorrection[AngleOffset10deg]"][index] = random.randint(30,60) 
    # IgnitionTiming[Angle]   
        df["IgnitionTiming[Angle]"][index] = random.randint(190.0,260.0)
    # Base Ignition
        df['BaseIgnition[AngleIgnSprt2K]'][index] = df["IgnitionTiming[Angle]"][index]+df["IgnitionCoolantTempCorrection[AngleOffset10deg]"][index]+df["IgnitionAirTempCorrection[Angle]"][index]
    
    df["cluster"] = -1    

    df.to_csv("../baddata/{}".format(file.split("\\")[-1]))
    
    if all_data is None:
        all_data = df
    else:
        all_data = pd.concat([all_data, df])

all_data.to_csv("../baddata/alldata.csv")
```

```python
import pandas as pd
import seaborn as sns
import glob
import matplotlib.pyplot as plt


input_files = glob.glob("../clean_data/ECU_DATA/2018*-home2mimos.csv", recursive=True)
trip = []
i=0
for file in input_files:
    if i==26:
        break
    values = []
    start = 2
    end = 750
    df = pd.read_csv(file)
    while(end < len(df[df.columns[0]])):
        values.append(df.iloc[start:end].mean())
        start += 750
        end += 750
    trip.append(values)
    i += 1
trip[0][0]
```

```python
trip_ignition = []
for day in range(len(trip)):
    day_airtemp = []
    for hr_df in trip[day]:
        j = 0
        for col in hr_df:
            if( j == 5):
                day_airtemp.append(col)
            j += 1
    trip_ignition.append(day_airtemp)

trip_sno = []
for day in range(len(trip)):
    day_rpm = []
    for hr_df in trip[day]:
        j = 0
        for col in hr_df:
            if( j == 1):
                day_rpm.append(col)
            j += 1
    trip_sno.append(day_rpm)


plt.plot(trip_sno[0],trip_ignition[0],label="day1")
plt.plot(trip_sno[1],trip_ignition[1],label="day2")
plt.plot(trip_sno[2],trip_ignition[2],label="day3")
plt.plot(trip_sno[20],trip_ignition[20],label="day20")
plt.plot(trip_sno[21],trip_ignition[21],label="day21")
# plt.plot([i for i in range(len(trip_ignition[1]))],list(trip_ignition[1]),label="day2")
# # plt.plot([i for i in range(len(trip_ignition[2]))],list(trip_ignition[2]),label="day3")
plt.legend()
plt.show()
```

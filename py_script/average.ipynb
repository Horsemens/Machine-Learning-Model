```python
import numpy as np
import pandas as pd
import seaborn as sns
import glob
import matplotlib.pyplot as plt
import os
import shutil

AVG_DATA_DIR_PATH = "../avg/"
ECU_DATA_DIR_PATH = AVG_DATA_DIR_PATH +"/ECU_DATA"

if(os.path.exists(AVG_DATA_DIR_PATH)):
    shutil.rmtree(AVG_DATA_DIR_PATH)


os.makedirs(ECU_DATA_DIR_PATH)

input_files = glob.glob("../clean_data/ECU_DATA/2018*.csv", recursive=True)
alldatadf = pd.DataFrame()
for file in input_files:
    df = pd.read_csv(file)
    df = df.groupby(np.arange(len(df))//750).mean()
    df.drop(columns=["Unnamed: 0", "sno"], inplace=True)
    df['sno'] = range(1, 1+len(df))
    if alldatadf is None:
        alldatadf = df
    else:
        alldatadf = pd.concat([alldatadf, df])
    df.to_csv("{}{}".format(AVG_DATA_DIR_PATH, file.split("/")[-1]))

alldatadf.drop(columns=['sno'], inplace=True)
alldatadf['sno'] = range(1, 1+len(alldatadf))
alldatadf.to_csv("{}{}".format(AVG_DATA_DIR_PATH, "all_data.csv"))
```

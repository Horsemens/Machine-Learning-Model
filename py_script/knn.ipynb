```python
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pickle as pkl
```

```python
df = pd.read_csv("../test_data.csv")
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python


features = [
    'IgnitionTiming[Angle]', 
    'MAPSource[Pressure]', 
    'RPM[EngineSpeed]', 
    'AFRDifference[AFR]', 
    'BatteryVoltage[BatteryVoltage]', 
    'AirTemp[Temperature]', 
    'CoolantTemp[Temperature]', 
    'LambdaSensor1[AFR]', 
    'BaseFuel[Percentage]', 
    'BaseIgnition[AngleIgnSprt2K]',
    'cluster'
]

X_features = [
    'IgnitionTiming[Angle]', 
    'MAPSource[Pressure]', 
    'RPM[EngineSpeed]', 
    'AFRDifference[AFR]', 
    'BatteryVoltage[BatteryVoltage]', 
    'AirTemp[Temperature]', 
    'CoolantTemp[Temperature]', 
    'LambdaSensor1[AFR]', 
    'BaseFuel[Percentage]', 
    'BaseIgnition[AngleIgnSprt2K]',
]

y_features = ['cluster']




df_feature = df[features]
x = df_feature[X_features]
x.head()
```

```python
y = df_feature[y_features]
y.head()
```

```python
scaler = preprocessing.MinMaxScaler().fit(x)
x_feature_scaled = scaler.transform(x)

x_feature_scaled
```

```python

X_train, X_test, y_train, y_test = train_test_split(x_feature_scaled, y, random_state = 0)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Accuracy with k=5", accuracy_score(y_test, y_pred)*100)
pkl.dump(knn,open('model.pkl','wb'))

data = {
            'IgnitionTiming[Angle]':  [144.20 , 118.37], 
            'MAPSource[Pressure]': [358.77, 367.81], 
            'RPM[EngineSpeed]': [1326.70, 1459.45], 
            'AFRDifference[AFR]': [12.0, 21.48], 
            'BatteryVoltage[BatteryVoltage]': [12885.4, 12922.14], 
            'AirTemp[Temperature]': [3075.52, 3304.68], 
            'CoolantTemp[Temperature]': [3604.35, 3598.62], 
            'LambdaSensor1[AFR]':[987, 980.87], 
            'BaseFuel[Percentage]':[359.55, 368.98], 
            'BaseIgnition[AngleIgnSprt2K]':[320.05, 295.68]
}
df_me = pd.DataFrame(data)

df_me_f = df_me[X_features]
scaler1 = preprocessing.MinMaxScaler().fit(df_me_f)
df_me_f_scaled = scaler1.transform(df_me_f)

print(knn.predict(df_me_f_scaled))
```
